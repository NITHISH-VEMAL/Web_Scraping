```python
!pip install bs4
from urllib.request import urlopen    #to open url
from bs4 import BeautifulSoup         #extract from html files
```

```python
url = "https://socializerofficial.000webhostapp.com/ques.php"  #my webpage url
html = urlopen(url)   #opens my webpage
```

```python
soup = BeautifulSoup(html, 'html')    #extracts and converts to python language format
type(soup)
```

```python
body = soup.body   #get the body of html
print(body)
type(body)
```

```python
text = soup.get_text()
print(text)   #can also print using for
```

```python
soup.find_all('tr')  #to find all the <tr> where tr is row of the table
```

```python
all_links = soup.find_all("a")   #links in my webpage
for link in all_links:    #traverse
    print(link.get("href"))
```

```python
rows = soup.find_all('td')    #from the <tr> now find <td> column
for i in rows:
    print(i)
    print()
```

```python
str_cells = str(rows)     #convert the rows to str format
cleantext = BeautifulSoup(str_cells, "html").get_text()    #excluding all the html tags such as <tr>, <td>, <form>
print(cleantext)
```

```python
#data refining


clean=[]
cleaned=[]
temp=[]
clean=cleantext.split(",")
for i in clean:
    if i=="    Show all comments":    
        cleaned.append(temp)
        temp=[]
    else:
        temp.append(i)
print(cleaned)
```

```python
#convert the list to dataframe format

df=pd.DataFrame(cleaned)
print(df)
```

```python
df[0] = df[0].str.split('says / asks', expand=True)   #split or strip the "says/ask" from the df
df[0]
```

```python
df[0] = df[0].str.strip('[')   #split the "[" from the first column
df.head(10)
```
